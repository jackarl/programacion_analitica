```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
N_total = 10000
t0 = 0.005  # 5 ms en segundos
c = 1e-4

# Función T(B)
def T(B):
    return (50 / B) + (1e-4 * B**2)

# Rango de B (evitar B=0)
B = np.linspace(1, 200, 500)
T_values = T(B)

# Encontrar el mínimo
min_idx = np.argmin(T_values)
B_min = B[min_idx]
T_min = T_values[min_idx]

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(B, T_values, label='$T(B) = \\frac{50}{B} + 10^{-4} B^2$')
plt.scatter(B_min, T_min, color='red', label=f'Mínimo en $B \\approx {int(B_min)}$')
plt.xlabel('Tamaño de lote $B$', fontsize=12)
plt.ylabel('Tiempo total $T(B)$ (s)', fontsize=12)
plt.title('Optimización del Tiempo de Entrenamiento dato pequeños', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Valor exacto
B_exact = (250000) ** (1/3)
print(f'Valor exacto de B: {B_exact:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros actualizados
N_total = 1_000_000
t0 = 1  # 1 segundo
c = 1e-4

# Función T(B)
def T(B):
    return (1_000_000 / B) + (1e-4 * B**2)

# Rango de B (evitar B=0)
B = np.linspace(500, 3000, 500)
T_values = T(B)

# Encontrar el mínimo
min_idx = np.argmin(T_values)
B_min = B[min_idx]
T_min = T_values[min_idx]

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(B, T_values, label='$T(B) = \\frac{1,\!000,\!000}{B} + 10^{-4} B^2$', color='blue')
plt.scatter(B_min, T_min, color='red', label=f'Mínimo en $B \\approx {int(B_min)}$', s=100)
plt.xlabel('Tamaño de lote $B$', fontsize=12)
plt.ylabel('Tiempo total $T(B)$ (s)', fontsize=12)
plt.title('Optimización del Tiempo de Entrenamiento (datos medianos)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()

# Valor exacto
B_exact = (5e9) ** (1/3)
print(f'Valor exacto de B: {B_exact:.2f}')
```
