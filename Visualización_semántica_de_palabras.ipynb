```python
from gensim.models import keyedvectors
import gensim.downloader as api
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
```

```python
### cargar modelo d emmbeddings en españo ###
model = api.load("fasttext-wiki-news-subwords-300")
print('Modelo instalado')
```

```python
palabra = ['futbol', 'ingeniero', 'cocina','mecanico','barcelona','baloncesto','cesto','carcel','politico','meta','felino','canino','animal','humano','rana','enfermeras','Doctor','hombre', 'mujer']
```

```python
palabras_filtradas = [p for p in palabra if p in model.key_to_index]
print("Palbras que usaremos:", palabras_filtradas)
```

```python
X = np.array([model[p] for p in palabras_filtradas])
print(X.shape)
print(X[0])
```

```python
# reduccion de dimensionalidad con t-SNE
tsne = TSNE(n_components=2,   random_state=42,  perplexity=5,   init="pca" )
X_tsne = tsne.fit_transform(X)
print(X_tsne.shape)
print(X_tsne)
```

```python
plt.figure(figsize=(10, 10))
for i, p in enumerate(palabras_filtradas):
  x,y = X_tsne[i, 0], X_tsne[i, 1]
  plt.scatter(x, y, c="blue")
  plt.text(x+0.05, y+0.05, p, fontsize=12)
  plt.show()
```
